```python
import json
import plotting as plt

with open("../../data/results/solution-quality.json", 'r', encoding='utf-8') as file:
    data_dict = json.load(file)
```

```python
x_values = [data_dict[x]["length_values"] for x in data_dict.keys()]
x_labels = [str(x) for x in data_dict.keys()]

plt.boxplot(x_values, x_labels, x_label="Number of Locations", y_label="Total Tour Length", title="Boxplot Tour Length")
```

```python
x_values = [data_dict[x]["z_values"] for x in data_dict.keys()]
x_labels = [str(x) for x in data_dict.keys()]

plt.boxplot(x_values, x_labels, x_label="Number of Locations", y_label="Total Tour Length", title="Boxplot Random Variate Z")
```
